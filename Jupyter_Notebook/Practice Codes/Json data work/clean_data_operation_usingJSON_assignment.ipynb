```python
import json
```

```python
def data_print(file_name):
    with open(file_name,"r") as f:
        f_data = json.load(f)
    
    return f_data
    
```

```python
j_data = data_print("store.json")
print(j_data)
```

```python
#Cleaning data
def clean_data(data):
    rating_to_num={"one" : 1,"two":2,"tree":3, "four":4, "five" : 5}
    cleaned_user=[]
    unq_user=set()
    for user in data:
        #cleaning rating
        raw_rating=str(user["rating"]).lower().strip()
        if (raw_rating in rating_to_num):
            raw_rating=rating_to_num[raw_rating]
        user["rating"]=raw_rating
        
        #eleaning null
        raw_age=user.get("age")
        if (raw_age==None):
            user["age"]=None

        #duplication fix
        if(user["name"].strip() in unq_user):
            continue
        unq_user.add(user["name"])
        cleaned_user.append(user)
    return cleaned_user
    
```

```python
data = clean_data(j_data)
print(data)
```

```python
def meaningful_data(data):
    total_rating = 0
    for user in data:
        total_rating+=float(user["rating"])
    print(f"avg rating {total_rating/len(data)}")

    poor_rating=0
    for user in data:
        if(float(user["rating"])<3):
            poor_rating+=1
    print(f"parcentage {poor_rating/len(data)*100}")
    
        
```

```python
meaningful_data(data)
```
